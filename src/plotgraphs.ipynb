```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
pd.plotting.register_matplotlib_converters()
```

```python
cache_data = pd.read_csv("../data/cache_data.csv",index_col="cache_size")
block_data = pd.read_csv("../data/block_data.csv",index_col="block_size")
associativity_data = pd.read_csv("../data/associativity_data.csv",index_col="associativity")
```

```python
for col in cache_data.columns:
    plt.figure(figsize=(12,6))
    sns.set_style("darkgrid")
    sns.lineplot(x = 'cache_size',y = 100 - cache_data[col],data = cache_data,marker ='o')
    plt.xticks([2**x for x in range(7,13)])
    plt.xlabel("cahce_size")
    plt.grid(True)
    plt.ylabel("miss_rate")
    plt.title("miss_rate% vs cache size " + col)
    plt.savefig("../graphs/"+"cache_"+col+'.png')
```

```python
plt.figure(figsize=(12,6))
sns.set_style("darkgrid")
plt.xticks([2**x for x in range(7,13)])
plt.xlabel('cache_Sizes')
plt.ylabel('miss rates%')
plt.title("miss rate% vs cache sizes for differentt traces")
plt.grid(True)
for col in cache_data.columns:
    sns.lineplot(x='cache_size',y=100-cache_data[col],data=cache_data,marker='o')
plt.legend(labels=['swim.trace','_','gzip.trace','_','gcc.trace','_','twolf.trace','_','mcf.trace'])
plt.savefig("../graphs/all_cache"+'.png')
```

```python
for col in block_data.columns:
    plt.figure(figsize=(12,6))
    sns.set_style("darkgrid")
    sns.lineplot(x = 'block_size',y = 100 - block_data[col],data = block_data,marker ='o')
    plt.xticks([2**x for x in range(1,8)])
    plt.xlabel("block_size")
    plt.grid(True)
    plt.ylabel("miss rate%")
    plt.title("miss rate% vs block size " + col)
    plt.savefig("../graphs/"+"block_"+col+'.png')
```

```python
for col in associativity_data.columns:
    plt.figure(figsize=(12,6))
    sns.set_style("darkgrid")
    sns.lineplot(x = 'associativity',y = col,data = associativity_data,marker ='o')
    min_val = associativity_data[col].min()
    max_val = associativity_data[col].max()
    # plt.yscale(max_val-min_val/6)
    # print(max_val-min_val)
    # plt.yticks([min_val +(x*((max_val-min_val)/6)) for x in range(-2,8) ] )
    plt.xticks([2**x for x in range(1,7)])
    plt.xlabel("associativity")
    plt.grid(True)
    plt.ylabel("hit rate")
    plt.title("HIT RATE% vs associativity " + col)
    plt.savefig("../graphs/"+"associativity_"+col+'.png')
```
